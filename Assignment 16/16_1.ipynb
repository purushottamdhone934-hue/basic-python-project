```python
import pandas as pd

df = pd.read_csv("insurance.csv")   # Replace with your file name if needed

print(df.head(10))
```

```python
# Shape
print("Shape of Dataset:")
print(df.shape)

# Information
print("\nDataset Information:")
print(df.info())

# Statistical Summary
print("\nStatistical Summary:")
print(df.describe())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
numerical = df.select_dtypes(include=['int64','float64']).columns
categorical = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(numerical)

print("\nCategorical Columns:")
print(categorical)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ['age','bmi','children','expenses']

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
cat_cols = ['sex','smoker','region']

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df)
    plt.title(f"Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df[['age','bmi','children','expenses']].corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Average Expenses:")
print(df['expenses'].mean())

print("\nMaximum Expenses:")
print(df['expenses'].max())

print("\nMinimum Expenses:")
print(df['expenses'].min())

print("\nAverage Expenses by Smoker:")
print(df.groupby('smoker')['expenses'].mean())
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x='smoker', y='expenses', data=df)
plt.title("Expenses vs Smoker")
plt.show()

plt.figure(figsize=(6,5))
sns.boxplot(x='sex', y='expenses', data=df)
plt.title("Expenses vs Sex")
plt.show()
```
