```python
import numpy as np

# Generate marks of 10 students × 5 subjects
marks = np.random.randint(30, 101, (10, 5))

print("Student Marks:")
print(marks)

# Total marks of each student
total = marks.sum(axis=1)
print("\nTotal Marks:")
print(total)

# Average marks of each student
average = marks.mean(axis=1)
print("\nAverage Marks:")
print(average)

# Highest and Lowest scorer
highest = np.argmax(total)
lowest = np.argmin(total)

print("\nHighest Scoring Student Index:", highest)
print("Marks:", marks[highest])

print("\nLowest Scoring Student Index:", lowest)
print("Marks:", marks[lowest])

# Overall class statistics
print("\nOverall Class Mean:", marks.mean())
print("Overall Class Standard Deviation:", marks.std())

# Top 3 students
top3 = np.argsort(total)[-3:][::-1]

print("\nTop 3 Student Indices:", top3)
print("Marks of Top 3 Students:")
print(marks[top3])
```
