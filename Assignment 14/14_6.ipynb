```python
import numpy as np

arr = np.random.randint(1, 101, (5, 5))

print("Original Array:\n", arr)

# Diagonal elements
print("Diagonal Elements:")
print(np.diag(arr))

# Elements greater than 50
print("Elements Greater Than 50:")
print(arr[arr > 50])

# Replace elements less than 30 with 0
arr[arr < 30] = 0

print("Modified Array:")
print(arr)
```
