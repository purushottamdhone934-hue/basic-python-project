```python
import numpy as np

arr = np.random.randn(6, 6)

print(arr)

print("Shape:", arr.shape)
print("Size:", arr.size)
print("Data Type:", arr.dtype)

print("Maximum Index:", np.unravel_index(np.argmax(arr), arr.shape))
print("Minimum Index:", np.unravel_index(np.argmin(arr), arr.shape))

print("Top-left 3x3 Matrix:")
print(arr[:3, :3])

# Convert negative values to positive
arr[arr < 0] = np.abs(arr[arr < 0])

print("Modified Matrix:")
print(arr)

print("Mean:", arr.mean())
```
